```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://dockerlabs.es"

respuesta = requests.get(url)

respuesta.encoding = "UTF-8" #Probe esto por que me mostraba caracteres mal puestos, y pense que estaba mal la codificacion, pero no soluciono
```

```python
print(respuesta.text)
```

```python
if respuesta.status_code == 200:
    soup = BeautifulSoup(respuesta.text, 'html.parser')

    # Con el filtro soup que creamos vamos a usar el metodo find_all, para filtrar tambien aquellos divs
    # que tengan una llamada a la accion onclick

    maquinas = soup.find_all('div', onclick = True)

    #Ahora queremos extraer el texto de estos divs, entonces vamos a usar un bucle for para recorrer la informacion

    for maquina in maquinas:
        
        onclick_text = maquina['onclick']  #para extraer la informacion que esta proxima al elemento onclick
        
        nombre_maquina = onclick_text.split('"')[1] #para poder obtener el nombre de cada maquina que esta entre las comillas simples

        print(nombre_maquina)
else:
    print(f"Hubo un error al hacer la peticion {respuesta.status_code}")
```
